## Remove capitalization / punctuations / numbers

```python
import string
import re
def remove_punct(text):
    text = text.lower()
    text = re.sub(r'\b(\w+)[-](\w+)\b', r'\1 \2', text)

    pattern = r'[^\w\s]'

    # Remove punctuation using re.sub()
    text = re.sub(pattern, '', text)
    
    
    #Remove numbers
    text = re.sub(r'\d', '', text)
    #replace "pi" with 'investigator'
    text = " ".join(
        "investigator" if word in ["pi"]
        else word
        for word in text.split())
    
    
    # Check and replace words ending with 'ment', 'ing'
    words = []
    for word in text.split():
        if word.endswith(('ment', 'ing')):
            # Remove the suffix
            #word = word[:-3] if word.endswith('ing')
            word=word
        words.append(word)
    
    # Join the words back into a single string
    text = " ".join(words)
    
    
    text = " ".join(
        word if word not in ["withdraw","withdrawn","withdrew"] else "withdrawal"
        for word in text.split()
    )
    
    text = " ".join(
        word if word not in ["md","doctor"] else "physician"
        for word in text.split()
    )
    
    

    text = " ".join(
        word if word not in ["subject", "participant","subjects","pts", "participants"] else "patient"
        for word in text.split()
    )

    text = " ".join(
        word if word not in ["poor", "slow", "inadequate", "insufficient"] else "low"
        for word in text.split()
    )

    text = " ".join(
        word if word not in ["funding","fundings"] else "fund"
        for word in text.split()
    )

    text = " ".join(
        word if word not in ["accrue"] else "accrual"
        for word in text.split()
    )


    text = " ".join(
        "investigator" if word in ["pi", "principal investigator", "primary investigator", "original investigator","promoter"]
        else word
        for word in text.split()
    )

    text = " ".join(
        "enrollment" if word in ["enroll", "recruit", "recruitment", "recruiting", "enrol","inclusion","enrolment","enrolled"]
        else word
        for word in text.split()
    )
    text = " ".join(
        "other" if word in ["others"]
        else word
        for word in text.split()
    )
    text = " ".join(
        "other" if word in ["others"]
        else word
        for word in text.split()
    )
    text = " ".join(
        "patient" if word in ["patients","subject","subjects","pt","pts"]
        else word
        for word in text.split()
    )
    text = text.replace('adverse event', 'ae')
    text = text.replace('side effect', 'ae')
    text = text.replace('side effects', 'ae')
    text = text.replace('adverse events', 'ae')

    return text

# Test the function
print(remove_punct("y'all Randomized but didn't receive stu'dy dru'g/s"))

data["raw_text"] = data["raw_text"].apply(lambda x: remove_punct(x))
```

```python
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('wordnet2022')


! cp -rf /usr/share/nltk_data/corpora/wordnet2022 /usr/share/nltk_data/corpora/wordnet # temp fix for lookup error.
```

```python
from nltk.corpus import wordnet
from nltk.corpus import brown

wordnet_map = {"N":wordnet.NOUN, 
               "V":wordnet.VERB, 
               "J":wordnet.ADJ, 
               "R":wordnet.ADV
              }
    
train_sents = brown.tagged_sents(categories='news')
t0 = nltk.DefaultTagger('NN')
t1 = nltk.UnigramTagger(train_sents, backoff=t0)
t2 = nltk.BigramTagger(train_sents, backoff=t1)

def pos_tag_wordnet(text, pos_tag_type="pos_tag"):
    """
        Create pos_tag with wordnet format
    """
    pos_tagged_text = t2.tag(text)
    
    # map the pos tagging output with wordnet output 
    pos_tagged_text = [(word, wordnet_map.get(pos_tag[0])) if pos_tag[0] in wordnet_map.keys() else (word, wordnet.NOUN) for (word, pos_tag) in pos_tagged_text ]
    return pos_tagged_text

data['combined_postag_wnet'] = data['stopwords_removed'].apply(lambda x: pos_tag_wordnet(x))
```

## Lemmatization w and w/o POS

```python
%time 

from nltk.stem import WordNetLemmatizer

def lemmatize_word(text):
    """
        Lemmatize the tokenized words
    """

    lemmatizer = WordNetLemmatizer()
    lemma = [lemmatizer.lemmatize(word, tag) for word, tag in text]
    return lemma

# Test without POS Tagging
lemmatizer = WordNetLemmatizer()

data['lemmatize_word_wo_pos'] = data['stopwords_removed'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
data['lemmatize_word_wo_pos'] = data['lemmatize_word_wo_pos'].apply(lambda x: [word for word in x if word not in stop])
data.head()
```

## Another version of lemmatized where the words that appears less than 10 times in total are filtered

```python
import pandas as pd
from collections import Counter

# Tokenize the text in the raw_text column
tokenized_words = data['stopwords_removed']

# Flatten the list of tokenized words
all_words = [word for sublist in tokenized_words for word in sublist]

# Count the frequency of each word
word_freq = Counter(all_words)

# Filter the word frequencies to include only words that appear more than 10 times
word_freq_filtered = {word: freq for word, freq in word_freq.items() if freq < 5}

# Create a DataFrame from the filtered word frequencies
word_freq_df = pd.DataFrame(list(word_freq_filtered.items()), columns=['Word', 'Frequency'])

# Sort the DataFrame by frequency
word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

# Reset the index
word_freq_df.reset_index(drop=True, inplace=True)

# Define a function to remove tokens contained in word_freq_df
def remove_filtered_tokens(tokens):
    return [token for token in tokens if token not in word_freq_df['Word'].values]

# Apply the function to create the filter_lemmatize_text column
data['filter_lemmatize_text'] = data['lemmatize_word_w_pos'].apply(remove_filtered_tokens)
data['filter_lemmatize_text'] = [' '.join(map(str, l)) for l in data['filter_lemmatize_text']]

```

```python
empty_rows = data['filter_lemmatize_text'] == ''
# Rplace empty values in 'filter_lemmatize_text' with values from 'lemmatize_text'
data.loc[empty_rows, 'filter_lemmatize_text'] = data.loc[empty_rows, 'lemmatize_text']
```
